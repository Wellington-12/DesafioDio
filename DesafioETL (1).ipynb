**Bibliotecas**

```python
import pandas as pd
import google.oauth2.service_account
```

**EXTRAÇÃO DOS DADOS**

```python
livro = pd.read_csv('Livros.csv')
```

```python
livro
```

```python
emprestimos = pd.read_csv('emprestimos.csv')
```

```python
emprestimos
```

**TRANSFORMAÇÃO DOS DADOS**

```python
resultado_df = pd.concat([livro, emprestimos], axis=1)
```

```python
resultado_df
```

```python
resultado_df.drop(columns=["ID_Livro"],inplace=True)
```

```python
resultado_df.head(5)
```

```python
resultado_df
```

```python
resultado_df.count()
```

```python
resultado_df = resultado_df[resultado_df.ID_Emprestimo > 0]
```

```python
resultado_df
```

**Carregamento dos dados**

```python
key_path = 'CHV.json'
```

```python
credentials = service_account.Credentials.from_service_account_file(key_path,scopes=['https://www.googleapis.com/auth/cloud-platform'])
```

```python
resultado_df.to_gbq(credentials=credentials,destination_table='Desafio_etl.Desafio',if_exists='replace')
```
